## Program for Forward Kinematics and inverse kinematics of 2R Planar Manipulator

üîß What is a 2R Planar Manipulator?

    2R means it has 2 Rotational (Revolute) joints.

    Planar means the motion is confined to 2D (X-Y plane).

üìê Link Configuration:

    Link 1: Length = L1, angle = Œ∏1

    Link 2: Length = L2, angle = Œ∏2 (relative to Link 1)

üßÆ Goal of Forward Kinematics:

    To calculate the end-effector position (x, y) given joint angles Œ∏‚ÇÅ and Œ∏‚ÇÇ and link lengths L‚ÇÅ and L‚ÇÇ.

‚úÖ Forward Kinematics Equations:



ax=L1cos‚Å°(Œ∏1)

ax=L1sin(Œ∏1)

Let‚Äôs derive the position of the end-effector (bx, by) or pose:

bx=L1cos‚Å°(Œ∏1)+L2cos‚Å°(Œ∏1+Œ∏2)

by=L1sin‚Å°(Œ∏1)+L2sin‚Å°(Œ∏1+Œ∏2)


üåÄ inverse kinematics:

Let‚Äôs derive Œ∏1,Œ∏2 from pose (bx,by)



```python
#imports

import numpy as np
import matplotlib.pyplot as plt

```

### Program for Forward Kinematics of 2R Planar Manipulator


```python
# Line for x-axis
x_axis_coordinate_x = [-10, 10]
x_axis_coordinate_y = [0, 0]

# Line for y-axis
y_axis_coordinate_x = [0, 0]
y_axis_coordinate_y = [-10, 10]


# links_ax
theta_1=np.radians(30)
a1=5

ax= a1*np.cos(theta_1)
ay= a1*np.sin(theta_1)
link_ax=[0,ax]
link_ay=[0,ay]

# links_bx
theta_2=np.radians(45)
a2=8

bx=ax+ a2*np.cos(theta_1+theta_2)
by=ay+ a2*np.sin(theta_1+theta_2)
link_bx=[ax,bx]
link_by=[ay,by]

plt.plot(x_axis_coordinate_x,x_axis_coordinate_y, color='black', )  # Horizontal line
plt.plot(y_axis_coordinate_x,y_axis_coordinate_y, color='black', )  # Vertical line
plt.plot(link_ax,link_ay, color='b',label='link_1',marker='o', linestyle='-',linewidth=3)
plt.plot(link_bx,link_by, color='r',label='link_2',marker='o', linestyle='-',linewidth=2)
plt.legend()
plt.grid(True, axis='both',color='gray',linestyle='--',linewidth=0.8,alpha=0.9)
plt.title("Forward Kinematics of 2R Planar Manipulator")

# To keep both axes proportionate
plt.show()
```

## ü§ñ What is a 2R Planar Manipulator?

    2R = 2 Revolute (rotational) joints

    Planar = Moves in 2D plane (like a robotic arm lying flat on a table)

It has:

    Two links: L‚ÇÅ and L‚ÇÇ

    Two joint angles: Œ∏‚ÇÅ and Œ∏‚ÇÇ

    End-effector located at (x, y) in 2D space

üéØ Goal of Inverse Kinematics

Given:

    Desired position of end-effector: (x, y)

    Link lengths: L‚ÇÅ, L‚ÇÇ

Find:Joint angles Œ∏‚ÇÅ, Œ∏‚ÇÇ such that the end-effector reaches (x, y)

```python

# Link lengths
a1 = 10
a2 = 7

# Desired end-effector position
x = 12
y = 5

# Calculate cos(Œ∏2) using the Law of Cosines
cos_theta2 = (x**2 + y**2 - a1**2 - a2**2) / (2 * a1 * a2)

# Check if the point is reachable
if abs(cos_theta2) > 1:
    raise ValueError("The point is outside the reachable workspace of the manipulator.")

# Elbow-down solution (use -acos(...) for elbow-up)
theta2 = np.arccos(cos_theta2)

# Compute Œ∏1 using geometric method
k1 = a1 + a2 * np.cos(theta2)
k2 = a2 * np.sin(theta2)
theta1 = np.arctan2(y, x) - np.arctan2(k2, k1)

# Convert to degrees
theta1_deg = np.degrees(theta1)
theta2_deg = np.degrees(theta2)

print(f"Joint Angle Œ∏1: {theta1_deg:.2f} degrees")
print(f"Joint Angle Œ∏2: {theta2_deg:.2f} degrees")

# Forward Kinematics for plotting
x0, y0 = 0, 0
x1 = a1 * np.cos(theta1)
y1 = a1 * np.sin(theta1)
x2 = x1 + a2 * np.cos(theta1 + theta2)
y2 = y1 + a2 * np.sin(theta1 + theta2)

# Plot
plt.figure()
plt.plot([x0, x1, x2], [y0, y1, y2], 'o-', linewidth=3)
plt.grid(True)
plt.axis('equal')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2R Planar Manipulator - Inverse Kinematics")
plt.show()
```

 ## What is a 3R Planar Manipulator?

    3R: It has 3 revolute (rotational) joints.

    Planar: All motion happens in a 2D plane (like a robot arm on a table).

    The links rotate, and the end-effector (tip of the arm) moves around in the plane.

üìç Problem Statement: Inverse Kinematics (IK)

Given:
The position and orientation of the end-effector (i.e., final x, y, and angle Œ∏).

Find:
The joint angles Œ∏‚ÇÅ, Œ∏‚ÇÇ, and Œ∏‚ÇÉ that achieve that pose.

| Step | What you do    | Formula                                                                             |
| ---- | -------------- | ----------------------------------------------------------------------------------- |
| 1    | Wrist Position | $x', y' = x - L‚ÇÉ\cos(œÜ), y - L‚ÇÉ\sin(œÜ)$                                             |
| 2    | Œ∏‚ÇÇ             | $\arctan2(¬±\sqrt{1 - D^2}, D)$, where $D = \frac{x'^2 + y'^2 - L‚ÇÅ^2 - L‚ÇÇ^2}{2L‚ÇÅL‚ÇÇ}$ |
| 3    | Œ∏‚ÇÅ             | $\arctan2(y', x') - \arctan2(L‚ÇÇ \sin(Œ∏‚ÇÇ), L‚ÇÅ + L‚ÇÇ \cos(Œ∏‚ÇÇ))$                        |
| 4    | Œ∏‚ÇÉ             | $œÜ - Œ∏‚ÇÅ - Œ∏‚ÇÇ$                                                                       |


```python
import numpy as np
import matplotlib.pyplot as plt

# Link lengths
a1 = 10
a2 = 8
a3 = 5

# Desired end-effector position and orientation
x = 15
y = 5
phi = np.radians(45)  # Desired orientation in radians

# Step 1: Wrist position
xw = x - a3 * np.cos(phi)
yw = y - a3 * np.sin(phi)

# Step 2: Œ∏2
cos_theta2 = (xw**2 + yw**2 - a1**2 - a2**2) / (2 * a1 * a2)
if abs(cos_theta2) > 1:
    raise ValueError("Unreachable point.")
theta2 = np.arccos(cos_theta2)

# Step 3: Œ∏1
k1 = a1 + a2 * np.cos(theta2)
k2 = a2 * np.sin(theta2)
theta1 = np.arctan2(yw, xw) - np.arctan2(k2, k1)

# Step 4: Œ∏3
theta3 = phi - theta1 - theta2

# Convert to degrees for display
theta1_deg = np.degrees(theta1)
theta2_deg = np.degrees(theta2)
theta3_deg = np.degrees(theta3)

# Print joint angles
print(f"Œ∏1 = {theta1_deg:.2f}¬∞")
print(f"Œ∏2 = {theta2_deg:.2f}¬∞")
print(f"Œ∏3 = {theta3_deg:.2f}¬∞")

# -------------------- PLOTTING --------------------

# Joint positions
x0, y0 = 0, 0  # base
x1 = a1 * np.cos(theta1)
y1 = a1 * np.sin(theta1)

x2 = x1 + a2 * np.cos(theta1 + theta2)
y2 = y1 + a2 * np.sin(theta1 + theta2)

x3 = x2 + a3 * np.cos(theta1 + theta2 + theta3)
y3 = y2 + a3 * np.sin(theta1 + theta2 + theta3)

# Create figure
fig, ax = plt.subplots()
ax.plot([x0, x1, x2, x3], [y0, y1, y2, y3], 'o-', linewidth=3, markersize=8)
ax.set_title('3R Planar Manipulator')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(-25, 25)
ax.set_ylim(-25, 25)
ax.set_aspect('equal', adjustable='box')
ax.grid(True)

# Annotate joints
ax.annotate("Base", (x0, y0), textcoords="offset points", xytext=(0,-10), ha='center')
ax.annotate("Joint 1", (x1, y1), textcoords="offset points", xytext=(0,-10), ha='center')
ax.annotate("Joint 2", (x2, y2), textcoords="offset points", xytext=(0,-10), ha='center')
ax.annotate("End-Effector", (x3, y3), textcoords="offset points", xytext=(0,-10), ha='center')

plt.show()
```
